```python
import pandas as pd
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pickle
```

```python
movies = pd.read_csv("../../data/processd/final_data.csv")
```

```python
movies.shape
```

```python
ps = PorterStemmer()
```

```python
def stemmer(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies['tags'] = movies['tags'].apply(stemmer)
```

```python
movies.iloc[0]['tags']
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(movies['tags']).toarray() # type: ignore
vector
```

```python
vector.shape
```

```python
similarity = cosine_similarity(vector)
similarity
```

```python
similarity.shape
```

```python
movies[movies['title'] == 'Spider-Man'].index[0]
```

```python
def recommend(movie):
    movie_index = movies[movies['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    for i in movies_list:
        print(movies.iloc[i[0]].title)
```

```python
recommend('Spider-Man')
```

```python
pickle.dump(movies, open('../../models/movies.pkl', 'wb'))
pickle.dump(similarity, open('../../models/similarity.pkl', 'wb'))
```
